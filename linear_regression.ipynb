```python
# remove outliers from age column using IQR method?
Q1_age = df_linear['age'].quantile(0.25)
Q3_age = df_linear['age'].quantile(0.75)
IQR_age = Q3_age - Q1_age
lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age
df_linear = df_linear[((df_linear['age'] >= lower_bound_age ) & (df_linear['age'] <= upper_bound_age))]
```

```python
X = df_linear[['age']]
y = df_linear['fare']
```

```python
sns.lmplot(data=df_linear, x='age', y='fare')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#model ko  import kare
model= LinearRegression()
 #model ko train  kare
model.fit(X,y)
#model ko predict karna ha
#for loop on model.predict function from 1 to 100?
for i in range(1, 101):  # loop from 1 to 100
    predictions = model.predict([[i]])  # make a prediction for each iteration
    print(f"Prediction for {i}: {predictions}")
# model.predict([[5]])

```

```python
#Model Evaluation With Metrics
X_train,  X_test, y_train, y_test = train_test_split(X, y, test_size=0.8)
model1= LinearRegression()
model1.fit(X_train, y_train)
predictions2=model1.predict(X_test)

```

```python
sns.scatterplot(x=y_test,y=predictions2)
```

```python
model1.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

MAE=mean_absolute_error(y_true=y_test,  y_pred=predictions2)
MSE=mean_squared_error(y_true=y_test,   y_pred=predictions2)
RMSE=mean_squared_error(y_true=y_test, y_pred=predictions2, squared=False)
R2=r2_score(y_true=y_test, y_pred=predictions2)

print("MAE:  ", MAE, "MSE: ",  MSE, "RMSE: ", RMSE, "R2: ", R2)

```
